## Visualize different physical observables vs temperature!
### magnetization: 
$$\langle \left|m\right|\rangle = \frac{1}{N} \langle
\left|\sum_i s_i\right|\rangle$$
### energy:
$$E = - J \sum_{\langle ij\rangle} s_i s_j - H \sum_i s_i$$
### specific heat:
$$C =
\frac{k_B \beta^2}{N} \left( \langle E^2 \rangle - \langle E \rangle^2
\right)
$$
### susceptibility:
$$\chi = \beta N \left(
\langle m^2 \rangle - \langle m \rangle^2 \right)$$

```python
temperature_grid, magnetization, susceptibility, specific_heat, energy = obs.get_obs_vs_temperature(
    cycles = 17000, 
    L = 32, 
    B = 0, 
    T_init = 2, 
    T_final = 3.1, 
    step = 0.1)
obs.plot_obs_vs_temperature(temperature_grid, magnetization, susceptibility, specific_heat, energy)
```
